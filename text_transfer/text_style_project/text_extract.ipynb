```python
import os
import re
import pandas as pd

def texts_to_sentences(txt_folder, char_len = 100):
    filenames = os.listdir(txt_folder)
    result = []
    books = []
    for name in filenames:
        filepath = txt_folder + '/' + name
        with open(filepath) as f:
            book = f.read()
        pattern = '[^.!?\n]{' + str(char_len) + ',}[.!?]' 
        sens = re.findall(pattern, book)
        result += sens
        books += [name] * len(sens)
    result_dict = {'text': result, 'book': books}
    sentences = pd.DataFrame(result_dict)
    sentences['text'] = sentences['text'].str.strip()
    sentences['text'] = sentences['text'].str.replace('^[^a-zA-Z“]+', '', regex=True)
    return sentences
        
```
